# Histograms on Iris Dataset
An histogram is an accurate graphical representation of the distribution of a numeric variable. The variable is cut into several bins, and the number of observation per bin is represented by the height of the bar

## Table of Contents:
- Histogram of Sepal Length (sepal length)
- Grid Histograms
- Histogram by value (Petal length by species)
- Horizontal Histogram

```python
## import package
import seaborn as sns
import matplotlib.pyplot as plt

## load data
df = sns.load_dataset("iris")
```

## Histogram of Sepal Length 

```python
## histogram
plt.figure(figsize=(12, 8))
sns.histplot(data=df, x="sepal_length")
plt.show()
```

## Grid histograms

```python
fig, axs = plt.subplots(2, 2, figsize=(12, 12))

sns.histplot(data=df, x="sepal_length", color="skyblue", ax=axs[0, 0])
sns.histplot(data=df, x="sepal_width", color="olive", ax=axs[0, 1])
sns.histplot(data=df, x="petal_length", color="gold", ax=axs[1, 0])
sns.histplot(data=df, x="petal_width", color="teal", ax=axs[1, 1])

plt.show()
```

## Histogram by value (petal length by species)

```python
sns.displot(df, x="petal_length", hue="species", element="step")
plt.show()
```

## Horizontal Histogram

```python
## set y to the values
plt.figure(figsize=(12, 8))
sns.histplot(data=df, y="sepal_length")
plt.show()
```

## KDE: Kernel Density Estimation

```python
plt.figure(figsize=(12, 8))
sns.displot(df, x="sepal_length", hue="species", kind="kde", multiple="stack")
plt.show()
```
